```python
from numpy import mean
from numpy import std
from matplotlib import pyplot as plt
from sklearn.model_selection import KFold
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.optimizers import SGD
from numpy import argmax
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.models import load_model
```

## Load Dataset

- we know that the images are all pre-aligned (e.g. each image only contains a hand-drawn digit), that the images all have the same square size of 28×28 pixels, and that the images are grayscale. Therefore, we can load the images and reshape the data arrays to have a single color channel.
- a one hot encoding for the class element of each sample, transforming the integer into a 10 element binary vector with a 1 for the index of the class value, and 0 values for all other classes

```python
def load_dataset():
  # load dataset
  (trainX, trainY), (testX, testY) = mnist.load_data()

  # reshape dataset to have a single channel
  trainX = trainX.reshape((trainX.shape[0], 28, 28, 1))
  testX = testX.reshape((testX.shape[0], 28, 28, 1))

  # one hot encode target values
  trainY = to_categorical(trainY)
  testY = to_categorical(testY)
  return trainX, trainY, testX, testY
```

## Prepare Pixel Data

- We know that the pixel values for each image in the dataset are unsigned integers in the range between black and white, or 0 and 255.

- We do not know the best way to scale the pixel values for modeling, but we know that some scaling will be required.

- A good starting point is to normalize the pixel values of grayscale images, e.g. rescale them to the range [0,1]. This involves first converting the data type from unsigned integers to floats, then dividing the pixel values by the maximum value.

```python
def prep_pixels(train, test):
  # convert from integers to floats
  train_norm = train.astype('float32')
  test_norm = test.astype('float32')

  # normalize to range 0-1
  train_norm = train_norm / 255.0
  test_norm = test_norm / 255.0

  # return normalized images
  return train_norm, test_norm
```

## Define Model



```python
def define_model():
  model = Sequential()
  model.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_uniform', input_shape=(28, 28, 1)))
  model.add(MaxPooling2D((2, 2)))
  model.add(Flatten())
  model.add(Dense(100, activation='relu', kernel_initializer='he_uniform'))
  model.add(Dense(10, activation='softmax'))

  # compile model
  opt = SGD(learning_rate=0.01, momentum=0.9)
  model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])

  return model
```

## Evaluate Model

```python
# evaluate a model using k-fold cross-validation
def evaluate_model(dataX, dataY, n_folds=5):
  scores, histories = list(), list()

  # prepare cross validation
  kfold = KFold(n_folds, shuffle=True, random_state=1)

  # enumerate splits
  for train_ix, test_ix in kfold.split(dataX):

    # define model
    model = define_model()

    # select rows for train and test
    trainX, trainY, testX, testY = dataX[train_ix], dataY[train_ix], dataX[test_ix], dataY[test_ix]

    # fit model
    history = model.fit(trainX, trainY, epochs=10, batch_size=32, validation_data=(testX, testY), verbose=0)

    # evaluate model
    _, acc = model.evaluate(testX, testY, verbose=0)
    print('> %.3f' % (acc * 100.0))

    # stores scores
    scores.append(acc)
    histories.append(history)

  return scores, histories
```

## Present Results

```python
def summarize_diagnostics(histories):
  for i in range(len(histories)):
    # plot loss
    plt.subplot(2, 1, 1)
    plt.title('Cross Entropy Loss')
    plt.plot(histories[i].history['loss'], color='blue', label='train')
    plt.plot(histories[i].history['val_loss'], color='orange', label='test')

    # plot accuracy
    plt.subplot(2, 1, 2)
    plt.title('Classification Accuracy')
    plt.plot(histories[i].history['accuracy'], color='blue', label='train')
    plt.plot(histories[i].history['val_accuracy'], color='orange', label='test')
    plt.show()
```

```python
def summarize_performance(scores):
  # print summary
  print('Accuracy: mean=%.3f std=%.3f, n=%d' % (mean(scores)*100, std(scores)*100, len(scores)))

  # box and whisker plots of results
  plt.boxplot(scores)
  plt.show()
```

```python
def run_test_harness():
  # load dataset
  trainX, trainY, testX, testY = load_dataset()

  # prepare pixel data
  trainX, testX = prep_pixels(trainX, testX)

  # evaluate model
  scores, histories = evaluate_model(trainX, trainY)

  # learning curves
  summarize_diagnostics(histories)

  # summarize estimated performance
  summarize_performance(scores)
```

```python
# entry point, run the test harness
run_test_harness()
```

## Save Final Model

```python
# run the test harness for evaluating a model
def run_test_harness2():
  # load dataset
  trainX, trainY, testX, testY = load_dataset()

  # prepare pixel data
  trainX, testX = prep_pixels(trainX, testX)
  # define model
  model = define_model()

  # fit model
  model.fit(trainX, trainY, epochs=10, batch_size=32, verbose=0)

  # save model
  model.save('final_model.h5')
```

## Extract label

```python
def extract_label(filename):
    label = int(filename.split('_')[0])
    return label
```

## Preprocess images

```python
import cv2
import numpy as np

def load_image(filename):
    img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)  # Đọc ảnh dưới dạng màu xám  [Original]
    img = cv2.GaussianBlur(img, (7,7), 0)  # Làm mịn ảnh bằng Gaussian Blur   [Blurred]

    img = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                cv2.THRESH_BINARY_INV, 7, 2)  # Áp dụng ngưỡng thích nghi để tạo ảnh nhị phân  [Thresholded]

    contours, _ = cv2.findContours(img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)  # Tìm contours trong ảnh
    # (Contours là các đường bao quanh một vùng liên tục có màu sắc hoặc cường độ sáng tương phản với vùng xung quanh)

    if contours:
        contour = max(contours, key=cv2.contourArea)  # Chọn contour lớn nhất      [With Contours]
        x, y, w, h = cv2.boundingRect(contour)  # Tính toán hình chữ nhật bao quanh contour
        img = img[y:y+h, x:x+w]  # Cắt ảnh theo hình chữ nhật bao    [Cropped]

    padding_size = max(img.shape) // 3  # Tính kích thước padding dựa trên kích thước ảnh

    img = cv2.copyMakeBorder(img, padding_size, padding_size, padding_size, padding_size, cv2.BORDER_CONSTANT, value=[0, 0, 0])  # Thêm padding vào ảnh để số nằm chính giữa và có độ lớn tương tự với ảnh train [Padded]

    img = cv2.resize(img, (28, 28), interpolation=cv2.INTER_AREA)  # Thay đổi kích thước ảnh thành 28x28 [Resized]

    img_array = img.astype('float32') / 255.0  # Chuẩn hóa ảnh về khoảng [0, 1]   [Normalized]

    img_array = img_array.reshape(1, 28, 28, 1)  # Chỉnh lại shape ảnh cho phù hợp với đầu vào mô hình

    return img_array
```

## Predict and calculate the accuracy of the model

```python
import os
def evaluate_example(directory):
    correct = 0
    total_img = 0
    model = load_model('final_model.h5')

    filenames = [f for f in sorted(os.listdir(directory)) if f.endswith('.png')]  # Truy xuất tất cả các tệp png từ thư mục
    total_files = len(filenames)

    # Tính toán số hàng cần thiết cho việc vẽ
    rows = total_files // 5 + (1 if total_files % 5 != 0 else 0)

    # Setup plot
    plt.figure(figsize=(25, rows * 5))

    for i, filename in enumerate(filenames):
        img_path = os.path.join(directory, filename)
        img = load_image(img_path)  # xử lý ảnh trước khi đưa vào model

        prediction = model.predict(img, verbose=0)
        predicted_class = argmax(prediction) # Lấy lớp có xác suất cao nhất

        actual_class = extract_label(filename) # Trích xuất lớp thực tế từ tên tệp

        if predicted_class == actual_class:
            correct += 1
        total_img += 1

        plt.subplot(rows, 5, i + 1) # # Đặt vị trí để vẽ ảnh

        processed_img_display = np.squeeze(img)
        plt.imshow(processed_img_display, cmap='gray') # Hiển thị ảnh
        plt.title(f'Actual: {actual_class} | Predicted: {predicted_class}')
        plt.axis('off')

    accuracy = correct / total_img if total_img > 0 else 0
    print(f"Total images: {total_img}, Model Accuracy: {accuracy*100:.2f}%")
    plt.tight_layout()
    plt.show()
```

```python
evaluate_example('/content/myimg/num_data')
```
